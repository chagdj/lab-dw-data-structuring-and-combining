```python
#Find Common Columns (Intersection):


columns_df1 = set(df1.columns)
columns_df2 = set(df2.columns)
columns_df3 = set(df3.columns)


common_columns = columns_df1 & columns_df2 & columns_df3
print("Common columns across all DataFrames:", common_columns)

```

```python
#Find Unique Columns for Each DataFrame:

unique_to_df1 = columns_df1 - columns_df2 - columns_df3
unique_to_df2 = columns_df2 - columns_df1 - columns_df3
unique_to_df3 = columns_df3 - columns_df1 - columns_df2

print("Unique to df1:", unique_to_df1)
print("Unique to df2:", unique_to_df2)
print("Unique to df3:", unique_to_df3)
```

```python

df1 = df1.rename(columns={'St': 'State'})
df2= df2.rename(columns={'St': 'State'})

```

```python
df = df3.merge(df1, on='Customer', how="left").merge(df2,on='Customer',how="inner").reset_index()
df 


```

```python
df.columns.nunique()
```

```python
columns = df.columns.tolist()
columns
```

```python
from collections import defaultdict

duplicates = defaultdict(list)

prefix_length = 5

for col in columns:
    prefix = col[:prefix_length]  # Extraire les premiers `prefix_length` caractères
    duplicates[prefix].append(col)

# Filter groups that have more than one word
result = {prefix: group for prefix, group in duplicates.items() if len(group) > 1}

for prefix, group in result.items():
    print(f"Prefix: {prefix}, Columns: {group}")




```

# Challenge 2: Structuring Data

In this challenge, we will continue to work with customer data from an insurance company, but we will use a dataset with more columns, called marketing_customer_analysis.csv, which can be found at the following link:

https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv

This dataset contains information such as customer demographics, policy details, vehicle information, and the customer's response to the last marketing campaign. Our goal is to explore and analyze this data by performing data cleaning, formatting, and structuring.

```python
import pandas as pd

# Load dataset directly from URL
url = "https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis_clean.csv"


df_custo  = pd.read_csv(url)

# Display first few rows
df_custo.head()
```

```python
df_custo.drop(columns="unnamed:_0", inplace=True)
```

```python
df_custo.head()
```

```python
df_custo.columns
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
sales_summary = df_custo.pivot_table(
    values='total_claim_amount', 
    index='sales_channel', 
    aggfunc='sum'
)

# Rounding to 2 decimal places
sales_summary = sales_summary.round(2)
sales_summary 
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
sales_channel_revenue = df_custo.pivot_table(values="total_claim_amount", index="sales_channel", aggfunc="sum").round(2)
print(sales_channel_revenue)

# 2. Pivot table for average customer lifetime value per gender and education level
clv_pivot = df_custo.pivot_table(values="customer_lifetime_value", index="gender", columns="education", aggfunc="mean").round(2)
print(clv_pivot)
```

## Bonus

You work at the customer service department and you want to know which months had the highest number of complaints by policy type category. Create a summary table showing the number of complaints by policy type and month.
Show it in a long format table.

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
complaints_pivot = df_custo.pivot_table(values="number_of_open_complaints", index=["month", "policy_type"], aggfunc="sum").reset_index()

complaints_pivot 
```

```python
long_format_complaints = df_custo.melt(id_vars=["month", "policy_type"], value_vars=["number_of_open_complaints"], var_name="metric", value_name="complaint_count")
long_format_complaints 
```
